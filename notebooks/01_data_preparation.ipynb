# Description des colonnes du dataset (Mobile Money Fraud)


## period
**Traduction :** Période / moment de la transaction  
**Explication :** Représente une dimension temporelle (temps simulé : heure, jour ou intervalle). Permet d’analyser les comportements dans le temps et de détecter des anomalies temporelles.

---

## operation
**Traduction :** Type d’opération  
**Explication :** Catégorie de la transaction (ex : transfert, retrait, dépôt, paiement). Certaines opérations peuvent être plus risquées que d’autres.

---

## amount
**Traduction :** Montant de la transaction  
**Explication :** Quantité d’argent transférée. C’est une variable clé pour détecter des comportements anormaux (montants très élevés ou inhabituels).

---

## origin_account
**Traduction :** Compte émetteur  
**Explication :** Identifiant du compte qui envoie l’argent. Permet d’analyser le comportement des utilisateurs et de détecter des comptes suspects.

---

## origin_balance_before
**Traduction :** Solde du compte émetteur avant transaction  
**Explication :** Montant disponible sur le compte avant l’opération. Utilisé pour vérifier la cohérence avec le montant transféré.

---

## origin_balance_after
**Traduction :** Solde du compte émetteur après transaction  
**Explication :** Montant restant après la transaction. Permet de calculer la variation du solde et détecter des incohérences.

---

## destination_account
**Traduction :** Compte receveur  
**Explication :** Identifiant du compte qui reçoit l’argent. Peut servir à détecter des réseaux de fraude ou des comptes impliqués dans plusieurs transactions suspectes.

---

## destination_balance_before
**Traduction :** Solde du compte receveur avant transaction  
**Explication :** Montant disponible sur le compte avant réception de l’argent.

---

## destination_balance_after
**Traduction :** Solde du compte receveur après transaction  
**Explication :** Montant après réception. Permet de vérifier la cohérence des flux financiers.

---

## fraud_flag
**Traduction :** Indicateur de fraude (variable cible)  
**Explication :** Variable à prédire :
- 0 = transaction normale  
- 1 = transaction frauduleuse  
Le modèle doit estimer la probabilité que cette valeur soit 1.
